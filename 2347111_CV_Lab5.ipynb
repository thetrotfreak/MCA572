## 1. Perform Two-Level Wavelet Transform

### Apply the 2D Discrete Wavelet Transform (DWT) to the given X-ray image to decompose it into frequency bands at two levels.

```python
# Import required libraries
import cv2
import pywt
import matplotlib.pyplot as plt

image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)

# Display the original X-ray image
plt.figure(figsize=(6, 6))
plt.imshow(image, cmap='gray')
plt.title('Original X-ray Image')
plt.show()

# Perform first-level 2D Discrete Wavelet Transform using Haar wavelet
coeffs1 = pywt.dwt2(image, 'haar')
LL1, (LH1, HL1, HH1) = coeffs1

# Perform second-level 2D Discrete Wavelet Transform on the LL1 band
coeffs2 = pywt.dwt2(LL1, 'haar')
LL2, (LH2, HL2, HH2) = coeffs2

# Display the first and second-level approximations (LL1 and LL2)
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.imshow(LL1, cmap='gray')

plt.title('First Level LL1 (Low-Low Approximation)')

plt.subplot(1, 2, 2)
plt.imshow(LL2, cmap='gray')
plt.title('Second Level LL2 (Low-Low Approximation)')

plt.show()

# Plot first-level decomposition (LL1, LH1, HL1, HH1)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# LL1: Low-frequency approximation
axs[0, 0].imshow(LL1, cmap='gray')
axs[0, 0].set_title('LL1 (Low-Low Approximation)')


# LH1: Vertical edges
axs[0, 1].imshow(LH1, cmap='gray')
axs[0, 1].set_title('LH1 (Low-High, Vertical Edges)')


# HL1: Horizontal edges
axs[1, 0].imshow(HL1, cmap='gray')
axs[1, 0].set_title('HL1 (High-Low, Horizontal Edges)')

# HH1: Diagonal details
axs[1, 1].imshow(HH1, cmap='gray')
axs[1, 1].set_title('HH1 (High-High, Diagonal Details)')


plt.show()

# Plot second-level decomposition (LL2, LH2, HL2, HH2)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# LL2: Further low-frequency approximation
axs[0, 0].imshow(LL2, cmap='gray')

axs[0, 0].set_title('LL2 (Low-Low Approximation)')


# LH2: Vertical features
axs[0, 1].imshow(LH2, cmap='gray')
axs[0, 1].set_title('LH2 (Low-High, Vertical Features)')


# HL2: Horizontal features
axs[1, 0].imshow(HL2, cmap='gray')
axs[1, 0].set_title('HL2 (High-Low, Horizontal Features)')


# HH2: Fine diagonal details
axs[1, 1].imshow(HH2, cmap='gray')
axs[1, 1].set_title('HH2 (High-High, Fine Details)')

plt.show()
```

## 2. Extract Frequency Bands:

### Decompose the image into the following frequency bands at both levels:
### Level 1 Bands:
- LL1 (Low-Low): Low-frequency approximation, representing coarse details of the image.
- LH1 (Low-High): Low-frequency in horizontal and high-frequency in vertical, emphasizing vertical edges.
- HL1 (High-Low): High-frequency in horizontal and low-frequency in vertical, emphasizing horizontal edges.
- HH1 (High-High): High-frequency details, capturing fine noise and sharp edges.
### Level 2 Bands (decompose the LL1 band further):
- LL2 (Low-Low): Further low-frequency approximation, providing a more refined view of coarse details.
- LH2 (Low-High): Low-frequency in horizontal and high-frequency in vertical, highlighting vertical features at a finer resolution.
- HL2 (High-Low): High-frequency in horizontal and low-frequency in vertical, emphasizing horizontal features at a finer resolution.
- HH2 (High-High): High-frequency details, capturing finer noise and sharper edges.

```python
# Perform two-level DWT on the image
coeffs1 = pywt.dwt2(image, 'haar')  # First level DWT
LL1, (LH1, HL1, HH1) = coeffs1  # Unpacking Level 1 coefficients

# Perform second-level DWT on LL1
coeffs2 = pywt.dwt2(LL1, 'haar')  # Second level DWT
LL2, (LH2, HL2, HH2) = coeffs2  # Unpacking Level 2 coefficients

# Display the extracted frequency bands for Level 1
plt.figure(figsize=(10, 10))

# LL1: Low-Low approximation
plt.subplot(2, 2, 1)
plt.imshow(LL1, cmap='gray')
plt.title('LL1 (Low-Low Approximation) - Level 1')
plt.axis('off')

# LH1: Low-High (Vertical edges)
plt.subplot(2, 2, 2)
plt.imshow(LH1, cmap='gray')
plt.title('LH1 (Low-High) - Level 1')
plt.axis('off')

# HL1: High-Low (Horizontal edges)
plt.subplot(2, 2, 3)
plt.imshow(HL1, cmap='gray')
plt.title('HL1 (High-Low) - Level 1')
plt.axis('off')

# HH1: High-High (Diagonal details)
plt.subplot(2, 2, 4)
plt.imshow(HH1, cmap='gray')
plt.title('HH1 (High-High) - Level 1')
plt.axis('off')

plt.tight_layout()
plt.show()

# Display the extracted frequency bands for Level 2
plt.figure(figsize=(10, 10))

# LL2: Low-Low approximation
plt.subplot(2, 2, 1)
plt.imshow(LL2, cmap='gray')
plt.title('LL2 (Low-Low Approximation) - Level 2')
plt.axis('off')

# LH2: Low-High (Vertical features)
plt.subplot(2, 2, 2)
plt.imshow(LH2, cmap='gray')
plt.title('LH2 (Low-High) - Level 2')
plt.axis('off')

# HL2: High-Low (Horizontal features)
plt.subplot(2, 2, 3)
plt.imshow(HL2, cmap='gray')
plt.title('HL2 (High-Low) - Level 2')
plt.axis('off')

# HH2: High-High (Fine details)
plt.subplot(2, 2, 4)
plt.imshow(HH2, cmap='gray')
plt.title('HH2 (High-High) - Level 2')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## 3. Display and Analyze Sub-Bands

### Display the resulting sub-bands from both levels. For each sub-band, provide a visual representation and a detailed explanation of what it represents

```python
# Perform two-level DWT using wavedec2
coeffs = pywt.wavedec2(image, 'haar', level=1)  # Perform DWT up to level 2

# Combine all coefficients into a single array for visualization
arr, coeff_slices = pywt.coeffs_to_array(
    coeffs)  # Get combined array of coefficients

# Plot the concatenated matrix of wavelet coefficients
plt.figure(figsize=(20, 20))
plt.imshow(arr, cmap=plt.cm.gray)
plt.title('All Wavelet Coefficients up to Level 1', fontsize=30)
plt.show()

interpretations = [
    "LL1: This is the low-frequency approximation, capturing the overall structure of the image with smooth features.",
    "LH1: Vertical edges are enhanced here, highlighting transitions and vertical details, such as the bones of the fingers.",
    "HL1: Horizontal edges are enhanced, making horizontal transitions more prominent, such as the wrist and fingers.",
    "HH1: Diagonal details are captured, focusing on finer structures like joints and diagonal transitions in the bones."
]

print("Interpretations of the Wavelet Coefficients:")
for i, interpretation in enumerate(interpretations):
    print(f"Level 1 - {i + 1}: {interpretation}")
```

## 4. Resolution Variation Analysis

### Compare the frequency bands at both levels to observe resolution variation
- Discuss how the details presented in LL2, LH2, HL2, and HH2 compare to their Level 1 counterparts.
- Analyze how increasing the level of decomposition affects the visibility of subtle features in the image.

```python
import matplotlib.pyplot as plt
import pywt
import numpy as np


# Perform two-level DWT using wavedec2
coeffs = pywt.wavedec2(image, 'haar', level=2)  # Perform DWT up to level 2

# Extract subbands for Level 1 and Level 2
LL2, (LH2, HL2, HH2) = coeffs[0], coeffs[1]
LL1, (LH1, HL1, HH1) = pywt.dwt2(LL2, 'haar')  # First level DWT again on LL2

# Get the sizes of each subband for display
sizes = {
    "LL1": LL1.shape,
    "LH1": LH1.shape,
    "HL1": HL1.shape,
    "HH1": HH1.shape,
    "LL2": LL2.shape,
    "LH2": LH2.shape,
    "HL2": HL2.shape,
    "HH2": HH2.shape,
}

# Create a figure to compare Level 1 and Level 2 subbands side by side with sizes
plt.figure(figsize=(16, 8))

# Plot Level 1 subbands
plt.subplot(2, 4, 1)
plt.imshow(LL1, cmap='gray')
plt.title(f'LL1 - Level 1\nSize: {sizes["LL1"]}')

plt.subplot(2, 4, 2)
plt.imshow(LH1, cmap='gray')
plt.title(f'LH1 - Level 1\nSize: {sizes["LH1"]}')


plt.subplot(2, 4, 3)
plt.imshow(HL1, cmap='gray')
plt.title(f'HL1 - Level 1\nSize: {sizes["HL1"]}')


plt.subplot(2, 4, 4)
plt.imshow(HH1, cmap='gray')
plt.title(f'HH1 - Level 1\nSize: {sizes["HH1"]}')


# Plot Level 2 subbands
plt.subplot(2, 4, 5)
plt.imshow(LL2, cmap='gray')
plt.title(f'LL2 - Level 2\nSize: {sizes["LL2"]}')


plt.subplot(2, 4, 6)
plt.imshow(LH2, cmap='gray')
plt.title(f'LH2 - Level 2\nSize: {sizes["LH2"]}')


plt.subplot(2, 4, 7)
plt.imshow(HL2, cmap='gray')
plt.title(f'HL2 - Level 2\nSize: {sizes["HL2"]}')


plt.subplot(2, 4, 8)
plt.imshow(HH2, cmap='gray')
plt.title(f'HH2 - Level 2\nSize: {sizes["HH2"]}')


plt.tight_layout()
plt.show()


text= """
Resolution Variation Analysis
"""
print(text)
print("Comparison of Levels 1 and 2: As you move from level 1 to level 2, the decomposition extracts finer and finer details, with LL2 showing a more blurred approximation compared to LL1. ")
print("The sub-bands at level 2 capture more refined edges and details, particularly for subtle features in the image.")
```
